```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## meist gesprochene Sprachen Zuhause im Zeitraum [2010, 2020]

```python
jahre = np.arange(2011, 2021)

df_sprachen_jahr = pd.read_excel("../Rohdaten/Die üblicherweise zu Hause gesprochenen Sprachen.xlsx", "2010", header=None) 
df_sprachen_jahr = df_sprachen_jahr.iloc[np.r_[5:18], np.r_[1:3]]
df_sprachen_jahr.columns = ["Sprache", "2010"]

for jahr in jahre:
    df_sp_zuh_jahr = pd.read_excel("../Rohdaten/Die üblicherweise zu Hause gesprochenen Sprachen.xlsx", str(jahr), header=None) 
    df_sp_zuh_jahr = df_sp_zuh_jahr.iloc[np.r_[5:18], 2]

    df_sprachen_jahr[jahr] = df_sp_zuh_jahr.values

df_sprachen_jahr.set_index("Sprache", inplace=True)
df_sprachen_jahr.columns.name = "Jahr"
df_sprachen_jahr
```

## heiraten nach staatsangehörigkeit der Ehepartner

```python
df_ehepartner_jahr = pd.read_excel("../Rohdaten/Heiraten nach Staatsangehörigkeitskategorie der Ehepartner, 1970-2020.xlsx", header=None) 
df_ehepartner_jahr = df_ehepartner_jahr.iloc[np.r_[4:55], np.r_[0:9]]

df_ehepartner_jahr.columns = ["Jahr", "Total_Schweizer_Schweizer", "Schweizer_Schweizer", "Total_Schweizer_Ausländer", "Ausländer_Schweizer", "Schweizer_Ausländer", "Total_Ausländer_Ausländer", "Auslänger_gleiche_Staatsangehörigkeit", "Auslänger_unterschiedliche_Staatsangehörigkeit"]
df_ehepartner_jahr.set_index("Jahr", inplace=True)
df_ehepartner_jahr = df_ehepartner_jahr[["Total_Schweizer_Schweizer", "Total_Schweizer_Ausländer", "Total_Ausländer_Ausländer"]]

df_ehepartner_jahr = df_ehepartner_jahr.transpose()
df_ehepartner_jahr.index.name = "Heiratstyp"
df_ehepartner_jahr
```

## first analysis

```python
df_compare = pd.concat([df_sprachen_jahr, df_ehepartner_jahr], join="inner")
df_compare = df_compare.transpose()

fig, ax = plt.subplots(1,1)

ax.plot(df_compare.index.values, df_compare["Schweizerdeutsch"].values, color="blue")

axy = ax.twinx()

axy.plot(df_compare.index.values, df_compare["Total_Schweizer_Schweizer"].values, color="green")
axy.plot(df_compare.index.values, df_compare["Total_Schweizer_Ausländer"].values, color="red")
axy.plot(df_compare.index.values, df_compare["Total_Ausländer_Ausländer"].values, color="purple")

plt.show()

df_compare.head()
```
